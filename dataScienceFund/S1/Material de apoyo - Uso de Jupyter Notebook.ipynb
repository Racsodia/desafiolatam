# Jupyter Notebook
- Es un entorno de trabajo para el desarrollo de la ciencia de datos, que funciona como una _aplicación web_. Su nombre viene de la unión de tres lenguajes de programación utilizados en la ciencia de datos: **Ju**lia, **Py**thon, y **R**. Su _kernel_ por tanto puede ser utilizado con cualquiera de estos 3 lenguajes, pero en este curso solo usaremos _el kernel de Python_.<br /><br />
- Esta aplicación se puede ejecutar como:<br /><br />
    - **Servidor local**: Forma más utilizada cuando se trabaja con pocos datos y/o datos cuyo uso no requiera de muchos recursos. _Es lo que utilizaremos en el curso_. <br /><br />
    - **Servidor en la nube**: Para exploraciones más costosas computacionalmente y/o cuando se requiere uso de GPU y ésta no está disponible en un servidor local. La más conocida es <a href="https://colab.research.google.com/">Google Colab</a>, que solo requiere de una cuenta de google (existen versiones de pago que permiten uso de más recursos). Otros servicios cloud, como Amazon Web Services (AWS), también permiten uso de Jupyter Notebook (en el caso de AWS, el servicio se llama Sagemaker).<br /><br />
    - **Servidor on premise**: En lugar de usar la máquina local como servidor, se utiliza un servidor (que es básicamente un computador con muchos recursos), pero que no está en "la nube", sino que en un computador al cual se tiene acceso físico.<br /><br />
    
- En este curso, la instalación de Jupyter Notebook es mediante <a href="https://www.anaconda.com/">Anaconda</a><br /><br />

- Para levantar el servidor web, se puede realizar mediante:
    - **Anaconda Navigator**: En Windows/Mac, buscar dentro de las aplicaciones instaladas. En Mac/Linux, inicializar con comando `anaconda-navigator` desde la terminal. Por defecto la aplicació iniciará en la _carpeta de usuario_.<br /><br />
    - **Línea de comandos**: En cualquier sistema operativo, ejecutando el comando `jupyter notebook` desde la línea de comandos o terminal. La aplicación iniciará en la _carpeta donde se ejecutó el comando_ en la terminal.<br /><br />
    
- Una vez inicializada la aplicación, no es posible "subir" en la jerarquía de carpetas "más arriba" desde donde se inicia la aplicación.<br /><br />
    
- Se crea un nuevo cuaderno con el botón "New" ubicado en la esquina superior derecha del árbol de navegación de archivos, y luego escogiendo el _kernel_, que en nuestro caso es "Python3".<br /><br />

- Los notebooks tienen "guardado automático" cada cierto tiempo, pero también se puede guardar el archivo explícitamente en la esquina superior izquierda, mediante el ícono de guardado (disquete).

# Celdas

- Para crear, cortar, copiar o pegar una celda, se debe utilizar los íconos ubicados en la parte superior izquierda (entre el ícono de "guardar" y las flechas de ordenamiento).
- Para mover una celda ya creada hacia arriba o hacia abajo, se debe usar las flechas direccionales.
- Para ejecutar una celda se debe utilizar el botón "Run", o la combinación de teclas `Shift` + `Enter`.
- Para detener la ejecución de una celda se debe utilizar el botón "Stop"
- Los dos tipos de celda que más utilizaremos son de tipo _Markdown_ y de tipo _Código_.

## Celda de Markdown
Markdown es un lenguaje de marcado muy utilizado que permite dar formato a texto. Permite uso de sintaxis en:
- Markdown como tal
- Html
- Latex
- Código en distintos lenguajes de programación

### Sintaxis de Markdown

Permite crear texto en **negrita**, _cursiva_.

Permite crear títulos y listas:
- # Título
- ## Subtítulo
- ### Sub sub título

Permite adjuntar imágenes 

Enlace a guía de Markdown: https://www.markdownguide.org/cheat-sheet/

### Sintaxis HTML

Permite el uso de toda sintaxis HTML, que es el lenguaje de hipertexto utilizado para dar formato visual a las páginas web. Esto es muy versátil ya que permite uso de imágenes, enlaces, listas, tablas, entre otros. Permite el uso de CSS (hojas de estilo), por medio del tag `style`.

### Sintaxis Latex
Permite uso de latex en línea envolviendo lo deseado entre '\\$'. También permite uso de bloques envolviendo lo deseado entre '$$'. Tiene la desventaja de que el uso de paquetes adicionales no puede ser incluído fácilmente.

Latex en línea: $\frac{1}{\sqrt(n)}$

Latex en bloque:
$$
\begin{align}
x &= a + b \\
&= 5 + 7 \\
&= 12
\end{align}
$$

### Sintaxis de código
Se debe utilizar envolviendo lo deseado entre tres \` . En la apertura, se debe añadir además el lenguaje del cual se quiere formatear la sintaxis.

Ejemplo bloque de Javascript
```javascript
console.log("Hola mundo")
```

Ejemplo bloque Python
```python
print("Hola mundo")
```
